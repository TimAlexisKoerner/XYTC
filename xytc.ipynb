```python
import matplotlib.pyplot as plt
from lattice import XYTCLattice

L = 6
xytc = XYTCLattice(L)

# Draw lines tracking diagonal configurations as seen in the paper
# Because the board is a standard axis-aligned grid, the diagonals
# run from corner to corner of the squares (x=0, y=0 to x=6, y=6)
mock_state_lines = [
    # A diagonal string across the lattice
    ((0, 0), (6, 6)),
    # An anti-diagonal string
    ((0, 4), (4, 0))
]

fig, ax = plt.subplots(figsize=(6, 6))
xytc.plot(ax=ax, state_lines=mock_state_lines)
plt.title(f"XYTC Chess Board Lattice (L={L})", fontsize=16)
plt.show()
```
